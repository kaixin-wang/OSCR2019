```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Data Import and Preprocessing

```python
# read in the data
data = pd.read_csv("Melbourne_housing_FULL.csv")
```

```python
# check if read in correctly
data.head()
```

```python
# remove entries with missing response value
data = data.loc[data.Price.notna()]
data = data.reset_index()
```

```python
# take log transformation to scale down the response
price = np.log(data.Price)
price = price.rename("logprice")
data = pd.concat([data, price], axis = 1)
data.head()
```

Exploratory Data Analysis

```python
# plot the boxplots for price
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 6))
fig.suptitle('Boxplots for price and log(price)')
ax1.boxplot(data.Price)
ax2.boxplot(data.logprice)
plt.show()
```

```python
plt.figure(figsize = (8, 6))
sns.heatmap(data.corr())
plt.title("correlation heatmap between variables")
plt.show()
```
